
## **Desenhando imagens e formas usando o OpenCV**


````python
import cv2
from matplotlib import pyplot as plt
import numpy as np

def imshow(image = None, size = 10):
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = h/w
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
````

````python
# Imagem preta usando o numpy para criar um array de zeros.
image = np.zeros((512,512,3), np.uint8)

# Imagem em escala cinza usando o numpy.
image_gray = np.zeros((512,512), np.uint8)

imshow(image)
imshow(image_gray)
````

### **Criando uma linha sobre nosso quadrado preto.**

```cv2.line(imagem, coordenadas de início, coordenadas de fim, cor, espessura)```

````python
# Observar que esta é uma operação direta, o que significa que ela altera a imagem 
# de entrada. Ao contrário de muitas outras funções do OpenCV, que retornam uma 
# nova imagem sem alterar a imagem de entrada. Lembre-se de que nossa imagem era 
# o quadro negro.
cv2.line(image, (0,0), (511,511), (255,127,0), 5)

imshow(image)
````

### **Criando Retângulos**

```cv2.rectangle(imagem, vértice inicial, vértice oposto, cor, espessura)```

````python
image = np.zeros((512,512,3), np.uint8)

cv2.rectangle(image, (100,100), (300,250), (127,50,127), -1)
imshow(image)
````

### **Agora vamos desenhar alguns círculos?**

```cv2.circle(imagem, centro, raio, cor, preenchimento)```

````python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (350, 350), 100, (15,150,50), -1) 
imshow(image)
````

### **Polígonos**

```cv2.polylines(imagem, pontos, Fechado?, cor, espessura)```

se Fechado = True, unimos o primeiro e último ponto.

**Observação**: cv2.polylines requer que nossos dados sejam organizados desta maneira:

````python
pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)
pts.shape
pts
````

````python
pts = pts.reshape((-1,1,2))
print(pts)
pts.shape
````

````python
image = np.zeros((512,512,3), np.uint8)

# 4 pontos
pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)

# Remodelamos nossos pontos na forma exigida pelo polylines.
pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts], True, (0,0,255), 3)
imshow(image)
````

### **Textos**

cv2.putText(imagem, 'Texto a ser exibido', ponto de início inferior esquerdo, Fonte, Tamanho da Fonte, Cor, Espessura)

Fontes disponíveis
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

````python
image = np.zeros((1000,1000,3), np.uint8)
ourString =  'Hello World!'
cv2.putText(image, ourString, (155,290), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 3, (40,200,0), 4)
imshow(image)
````
