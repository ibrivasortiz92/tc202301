```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

def imshow(image = None, size = 10):
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = h/w
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

## **Operações Aritméticas**

Operações simples que nos permitem adicionar ou subtrair diretamente a intensidade da cor.

```python
# Adicionar a vírgula zero no cv2.imread carrega nossa imagem como uma imagem em escala de cinza.
image = cv2.imread('images/liberty.jpeg', 0)
print(image.shape)
imshow(image)

M = np.ones(image.shape, dtype = "uint8") * 100 

added = cv2.add(image, M) # Not clip
imshow(added)

added2 = image + M 
imshow(added2)

print(image.dtype)
```

#### **Diminuindo o Brilho**

```python
subtracted = cv2.subtract(image, M)
imshow(subtracted)

subtracted = image - M 
imshow(subtracted)
```

## **Operações Bitwise**

```python

# Quadrado
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
imshow(square)

# Ellipse
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
imshow(ellipse)
```

```python
# Mostra apenas onde elas se intersectam.
And = cv2.bitwise_and(square, ellipse)
imshow(And)

```

```python
# Mostra onde está o quadrado ou a elipse.
bitwiseOr = cv2.bitwise_or(square, ellipse)
imshow(bitwiseOr)

```

```python
# Mostra onde cada um existe individualmente.
bitwiseXor = cv2.bitwise_xor(square, ellipse)
imshow(bitwiseXor)

```

```python
# Mostra tudo que não faz parte do quadrado.
bitwiseNot_sq = cv2.bitwise_not(square)
imshow(bitwiseNot_sq)
```
