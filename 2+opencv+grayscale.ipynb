
# **Convertendo imagens para escala de cinza.**

Tópicos:
1. Converter uma imagem colorida para escala de cinza.
2. Ver a diferença de dimensões entre imagens em escala de cinza e coloridas.


```python
import cv2
from matplotlib import pyplot as plt

def imshow(image = None, size = 10):
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = h/w
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# Imagem para teste.
image = cv2.imread('./images/castara.jpeg')

imshow(image)
```

```python
# Usamos o cvtColor para converter para escala de cinza.
# Recebe 2 argumentos, o primeiro sendo a imagem de entrada
# O segundo é o código de conversão do espaço de cores.
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow(gray_image)
```

### **Dimensões da imagem em escala de cinza.**

Imagens coloridas em RGB têm 3 dimensões, uma para cada cor primária. A escala de cinza possui apenas 1, que é a intensidade do cinza.
