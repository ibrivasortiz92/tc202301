```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

def imshow(image = None, size = 10):
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = h/w
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

- **Dilatação** - Adiciona pixels às bordas dos objetos em uma imagem.
- **Erosão** - Remove pixels nas bordas dos objetos em uma imagem.
- **Abertura** - Erosão seguida de dilatação.
- **Fechamento** - Dilatação seguida de erosão.

![](images/de.png)

```python

image = cv2.imread('images/opencv_inv.png', 0)
imshow(image)

kernel = np.ones((5,5), np.uint8)

# Erosao
erosion = cv2.erode(image, kernel, iterations = 1)
imshow(erosion)

# Dilatacao
dilation = cv2.dilate(image, kernel, iterations = 1)
imshow(dilation)

# Abertura
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
imshow(opening)

# Fechadura
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
imshow(closing)
```

## **Detecção de Bordas com Canny** 
![](images/canny.png)

- O primeiro argumento é a nossa imagem de entrada.
- O segundo e o terceiro argumentos são o minVal e o maxVal, respectivamente.
- O quarto argumento é o tamanho do aparelho. É o tamanho do kernel usado para encontrar os gradientes da imagem. Por padrão, é 3.

A detecção de bordas precisa de um limiar para determinar qual diferença/mudança deve ser considerada como borda.

```python
image = cv2.imread('images/londonxmas.jpeg',0)
imshow(image)

canny = cv2.Canny(image, 50, 120)
imshow(canny)

# Limiares amplos de borda são adequados quando há muitas bordas esperadas.
canny = cv2.Canny(image, 10, 200)
imshow(canny)

# Limiar estreito, espera-se menos bordas.
canny = cv2.Canny(image, 200, 240)
imshow(canny)

canny = cv2.Canny(image, 60, 110)
imshow(canny)
```
