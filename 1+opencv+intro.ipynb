OpenCV (Open Source Computer Vision Library) is an open-source computer vision and machine learning software library. It provides a wide range of functions and algorithms that enable developers to process images and videos, perform object detection and recognition, track motion, and much more. OpenCV is widely used in various fields, including robotics, augmented reality, surveillance, and image/video editing.

```python
# Importando OpenCV.
import cv2

cv2.__version__
```

### **Carregando imagens**

```python

# Carregue uma imagem usando 'imread', especificando o caminho
# da imagem.

image = cv2.imread('./images/castara.jpeg')
```

### **Exibindo imagens**

```python
from matplotlib import pyplot as plt

# Mostrar a imagem com o matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def imshow(image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
imshow(image)
```

### **Salvando imagens**

```python

# Simplesmente use 'imwrite', especificando o nome do arquivo
# e a imagem a ser salva.

cv2.imwrite('output.jpg', image)
```

```python

# Ou salve-a como PNG (Portable Network Graphics), 
# que é um formato de imagem bitmap sem perda de qualidade.

cv2.imwrite('output.png', image)
```

### **Exibindo as dimensões da imagem.**

Representação de imagens como arrays:

![](images/array.png)



```python
print(image.shape)
```

A primeira dimensão é a altura e tem 1280 pixels.

A segunda dimensão é a largura, que possui 960 pixels.

Podemos imprimi-las de forma organizada assim:

```python
print('Altura da Imagem: {} pixels'.format(int(image.shape[0])))
print('Largura da Imagem: {} pixels'.format(int(image.shape[1])))
print('Profundidade da Imagem: {} componentes de cores'.format(int(image.shape[2])))
```
