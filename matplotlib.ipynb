## 3. Gráficos Básicos

### Gráficos de Linha

Gráficos de linha são um tipo comum de visualização usado para exibir tendências ou padrões ao longo do tempo. Eles são particularmente úteis para representar dados contínuos, como preços de ações, variações de temperatura ou crescimento populacional.

No Matplotlib, podemos criar um gráfico de linha usando a função `plot()` do módulo `pyplot`. A função `plot()` recebe dois argumentos: os valores de x e os valores de y dos pontos de dados a serem plotados. Aqui está um exemplo básico:



````python
# Dados
x = [1, 2, 3, 6]
y = [2, 4, 6, 20]

# Criar um gráfico de linha
plt.plot(x, y)

# Exibir o gráfico
plt.show()
````


Neste exemplo, definimos duas listas, `x` e `y`, para representar os valores de x e y dos nossos pontos de dados, respectivamente. Em seguida, passamos essas listas como argumentos para a função `plot()`. Por fim, usamos `plt.show()` para exibir o gráfico.

Por padrão, o Matplotlib conecta os pontos de dados com linhas retas. No entanto, você pode personalizar a aparência da linha adicionando argumentos opcionais à função `plot()`. Por exemplo, você pode especificar a cor da linha, o estilo da linha e o estilo do marcador:

````python

plt.plot(x, y, color='red', linestyle='dashed', marker='o')
plt.show()
````

### Gráficos de Dispersão

Gráficos de dispersão são comumente usados para visualizar a relação entre duas variáveis. Eles são particularmente úteis para identificar padrões, agrupamentos ou correlações nos dados. Cada ponto de dados em um gráfico de dispersão é representado por um marcador, permitindo visualizar a distribuição e dispersão dos dados.

No Matplotlib, podemos criar gráficos de dispersão usando a função `scatter()` do módulo `pyplot`. A função `scatter()` recebe dois argumentos: os valores de x e os valores de y dos pontos de dados a serem plotados.


````python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criar um gráfico de dispersão
plt.scatter(x, y)

# Exibir o gráfico
plt.show()
````


Neste exemplo, definimos duas listas, `x` e `y`, para representar os valores de x e y dos nossos pontos de dados, respectivamente. Em seguida, passamos essas listas como argumentos para a função `scatter()`. Por fim, usamos `plt.show()` para exibir o gráfico.

Por padrão, o Matplotlib usa marcadores circulares para gráficos de dispersão. No entanto, você pode personalizar a aparência dos marcadores adicionando argumentos opcionais à função `scatter()`. Por exemplo, você pode especificar o tamanho do marcador, a cor do marcador e o estilo do marcador:


````python
plt.scatter(x, y, s=50, c='red', marker='o')
plt.show()
````

### Gráficos de Barras

Gráficos de barras são comumente usados para comparar categorias ou grupos, representando os dados como barras retangulares de alturas variadas. Eles são eficazes na visualização de dados categóricos e na exibição da magnitude ou frequência de diferentes grupos.

No Matplotlib, podemos criar gráficos de barras usando a função `bar()` do módulo `pyplot`. A função `bar()` recebe dois argumentos: os valores de x e as alturas correspondentes das barras.



````python
# Dados
categorias = ['A', 'B', 'C', 'D']
alturas = [10, 7, 5, 12]

# Criar um gráfico de barras
plt.bar(categorias, alturas)

# Exibir o gráfico
plt.show()
````


Neste exemplo, definimos duas listas, `categorias` e `alturas`, para representar as categorias ou grupos e as alturas das barras, respectivamente. Em seguida, passamos essas listas como argumentos para a função `bar()`. Por fim, usamos `plt.show()` para exibir o gráfico.

Por padrão, o Matplotlib cria barras verticais. No entanto, você pode criar barras horizontais usando a função `barh()` em vez disso. Por exemplo:


````python
plt.barh(categorias, alturas, height=0.1)
plt.show()
````


Você pode personalizar a aparência das barras adicionando argumentos opcionais à função `bar()` ou `barh()`. Por exemplo, você pode especificar a cor da barra, cor da borda e largura:


````python
plt.bar(categorias, alturas, color='blue', edgecolor='black', width=0.5)

plt.show()
````

## 4. Personalização dos Gráficos

Personalizar os gráficos nos permite aprimorar a apresentação visual dos nossos dados e tornar nossos gráficos mais informativos e atraentes visualmente.


### Anatomia de uma Figura

Antes de mergulharmos na personalização dos gráficos, vamos revisar brevemente os diferentes componentes de uma figura do Matplotlib. Compreender a anatomia de uma figura nos ajudará a identificar as partes que podemos personalizar.

Uma figura do Matplotlib é composta por vários componentes, incluindo:

- Figure: O contêiner geral que contém todos os elementos de um gráfico.
- Axes: A área onde os dados são plotados. Uma figura pode conter um ou mais eixos.
- Axis: O eixo x e o eixo y, que representam as escalas e rótulos para os dados plotados.
- Title: O título do gráfico.
- Labels: Os rótulos para o eixo x e o eixo y.
- Legends: Rótulos descritivos associados aos dados plotados.



````python
# Criar um gráfico simples
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])

# Adicionar rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Exemplo de Personalização de Gráficos')

# Exibir o gráfico
plt.show()
````

### Personalizando Eixos e Rótulos

Para personalizar a aparência dos eixos, temos várias opções à nossa disposição:

- Alterar Limites dos Eixos: Podemos modificar o intervalo de valores exibidos no eixo x e no eixo y usando `plt.xlim()` e `plt.ylim()`, respectivamente. Isso é útil quando queremos ampliar uma região específica do gráfico.

- Modificar Marcas: As marcas são os valores exibidos ao longo dos eixos. Podemos personalizar a aparência das marcas usando funções como `plt.xticks()` e `plt.yticks()`. Por exemplo, podemos alterar as posições, rótulos e formatação das marcas.

- Adicionar Linhas de Grade: Linhas de grade podem melhorar a legibilidade de um gráfico. Podemos habilitar as linhas de grade usando `plt.grid()`. Por padrão, as linhas de grade são exibidas tanto para o eixo x quanto para o eixo y, mas podemos personalizar esse comportamento.

- Rótulos dos Eixos: Podemos modificar os rótulos do eixo x e do eixo y usando `plt.xlabel()` e `plt.ylabel()`, respectivamente. Isso nos permite fornecer descrições significativas dos dados plotados.

- Título do Gráfico: O título de um gráfico pode ser definido usando `plt.title()`. Um bom título deve resumir o conteúdo do gráfico e transmitir a mensagem principal.


````python
# Criar um gráfico simples
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
plt.plot(x, y)

# Personalizar os eixos
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xticks([1, 2, 3, 4, 5])  # Locais personalizados das marcas
plt.yticks([0, 2, 4, 6])  # Rótulos personalizados das marcas
plt.grid(True)  # Habilitar linhas de grade

# Personalizar rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Exemplo de Personalização de Gráficos')

# Exibir o gráfico
plt.show()
````

### Estilizando Gráficos

Além de personalizar os eixos e rótulos, o Matplotlib fornece diferentes estilos embutidos que podem ser aplicados aos gráficos. Os estilos controlam vários aspectos da aparência do gráfico, como cores, estilos de linha e fontes. Usar um estilo predefinido pode ajudar a obter uma aparência consistente e profissional em vários gráficos.

O Matplotlib oferece uma variedade de estilos embutidos, incluindo 'default', 'classic', 'ggplot', 'seaborn' e outros. Para aplicar um estilo, utilizamos a função `plt.style.use()`.



````python
# Aplicar o estilo 'ggplot'
plt.style.use('seaborn-v0_8-bright')

# Criar e personalizar um gráfico
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
plt.plot(x, y)

# Exibir o gráfico
plt.show()
````

Neste exemplo, aplicamos o estilo 'ggplot', que fornece uma aparência limpa e moderna ao gráfico. Você pode experimentar diferentes estilos e escolher aquele que melhor se adapta às suas necessidades. Além disso, é possível criar estilos personalizados modificando aspectos específicos do gráfico, como cores, estilos de linha ou tamanhos de fonte.

### Adicionando Legendas

Legendas são um componente essencial dos gráficos, pois fornecem informações sobre os diferentes elementos ou categorias representadas no gráfico. No Matplotlib, podemos adicionar legendas aos nossos gráficos para fornecer clareza e melhorar a compreensão.

Para adicionar uma legenda a um gráfico, primeiro precisamos atribuir um rótulo a cada elemento que desejamos que apareça na legenda. Por exemplo, se tivermos várias linhas em um gráfico de linhas, atribuímos um rótulo a cada linha. Da mesma forma, em um gráfico de dispersão ou um gráfico de barras, atribuímos rótulos a diferentes grupos ou categorias.


````python
# Dados
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Criar gráficos de linhas
plt.plot(x, y1, label='Conjunto de Dados 1')
plt.plot(x, y2, label='Conjunto de Dados 2')

# Adicionar uma legenda
plt.legend(loc="upper left")

# Exibir o gráfico
plt.show()
````

Neste exemplo, criamos dois gráficos de linhas usando a função `plot()` e atribuímos rótulos a cada linha usando o parâmetro `label`. A função `legend()` é chamada para adicionar a legenda ao gráfico.


## 5. Salvando Gráficos

O Matplotlib nos permite salvar nossos gráficos como arquivos de imagem em vários formatos, como PNG, JPEG, PDF ou SVG. Essa funcionalidade é útil quando queremos usar o gráfico em outros documentos, relatórios ou apresentações.

Para salvar um gráfico, podemos usar a função `savefig()` do módulo `pyplot`. A função `savefig()` recebe o nome de arquivo desejado como argumento, juntamente com um parâmetro opcional para especificar o formato do arquivo.


````python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criar um gráfico de linhas
plt.plot(x, y)

# Salvar o gráfico como um arquivo PNG
plt.savefig('line_plot.jpg')
````

Neste exemplo, criamos um gráfico de linhas usando a função `plot()`. Em seguida, usamos a função `savefig()` para salvar o gráfico como um arquivo PNG chamado `line_plot.png`. Por padrão, o Matplotlib salva o gráfico no diretório de trabalho atual.

Você pode especificar o formato de arquivo desejado fornecendo a extensão do arquivo no nome do arquivo. Por exemplo, para salvar o gráfico como um arquivo JPEG, você pode usar a extensão `.jpg` ou `.jpeg`:

````python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criar um gráfico de linhas
plt.plot(x, y)

# Salvar o gráfico como um arquivo JPG
plt.savefig('line_plot.jpg')
````

O Matplotlib determina automaticamente o formato do arquivo com base na extensão fornecida. Se você não especificar o formato, o Matplotlib usará o formato especificado no parâmetro de configuração `savefig.format`.

Além do nome do arquivo, a função `savefig()` fornece vários parâmetros opcionais para controlar a saída. Por exemplo, você pode especificar a resolução (pontos por polegada) para a imagem salva usando o parâmetro `dpi`:


````python
# Dados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criar um gráfico de linhas
plt.plot(x, y)

# Salvar o gráfico como um arquivo PNG e 300 PDI
plt.savefig('line_plot.png', dpi=300)
````

## 6. Múltiplos Gráficos e Subgráficos

Trabalhar com múltiplos gráficos nos permite comparar e visualizar diferentes aspectos de nossos dados simultaneamente.


### Criando Múltiplos Gráficos

Para criar múltiplos gráficos em uma única figura, podemos usar a função `plt.subplot()`. Essa função recebe três argumentos: o número de linhas, o número de colunas e o índice do gráfico atual.


````python
# Criar o primeiro gráfico
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Gráfico 1')

# Criar o segundo gráfico
plt.subplot(1, 2, 2)
plt.plot([1, 2, 3, 4], [3, 2, 4, 1])
plt.title('Gráfico 2')

# Ajustar o layout
plt.tight_layout()

# Exibir os gráficos
plt.show()
````

Neste exemplo, usamos `plt.subplot(1, 2, 1)` para criar o primeiro gráfico à esquerda e `plt.subplot(1, 2, 2)` para criar o segundo gráfico à direita. O `1` no primeiro argumento especifica que queremos uma linha de gráficos, e o `2` no segundo argumento indica que queremos duas colunas de gráficos. O último argumento, `1` e `2`, especifica o índice do gráfico atual.


Podemos estender esse conceito para criar grades de gráficos com várias linhas e colunas. Por exemplo, para criar uma grade 2x2 de gráficos:


````python
# Criar o primeiro gráfico em uma grade 2x2
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Gráfico 1')

# Criar o segundo gráfico em uma grade 2x2
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3, 4], [3, 2, 4, 1])
plt.title('Gráfico 2')

# Criar o terceiro gráfico em uma grade 2x2
plt.subplot(2, 2, 3)
plt.plot([1, 2, 3, 4], [2, 3, 1, 4])
plt.title('Gráfico 3')

# Criar o quarto gráfico em uma grade 2x2
plt.subplot(2, 2, 4)
plt.plot([1, 2, 3, 4], [4, 1, 3, 2])
plt.title('Gráfico 4')

# Ajustar o layout
plt.tight_layout()

# Exibir os gráficos
plt.show()
````

### Compartilhando Eixos e Criando Inserções

Em alguns casos, pode ser útil compartilhar os eixos entre gráficos para sincronizar suas escalas, marcações e rótulos. Isso é especialmente útil ao comparar dados relacionados. O Matplotlib fornece a função `plt.subplots()`, que nos permite criar subgráficos que compartilham o eixo x, o eixo y ou ambos.


````python
# Criar dois subgráficos compartilhando o eixo x
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plotar dados no primeiro subgráfico
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax1.set_title('Gráfico 1')

# Plotar dados no segundo subgráfico
ax2.plot([1, 2, 3, 4], [3, 2, 4, 1])
ax2.set_title('Gráfico 2')

# Exibir os gráficos
plt.tight_layout()
plt.show()
````

Neste exemplo, criamos dois subgráficos usando `plt.subplots(2, 1, sharex=True)`. O argumento `sharex=True` garante que os subgráficos compartilhem o eixo x. Dessa forma, os limites, as marcações e os rótulos do eixo x são sincronizados entre os gráficos.


Outra técnica que podemos usar ao trabalhar com múltiplos gráficos é criar inserções. As inserções são gráficos menores colocados dentro de um gráfico maior. Elas são úteis para destacar regiões específicas ou fornecer informações adicionais relacionadas ao gráfico principal.


````python
# Criar o gráfico principal
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.title('Gráfico Principal')

# Criar a inserção
inset_axes = plt.axes([0.2, 0.6, 0.25, 0.25])  # [esquerda, inferior, largura, altura]
inset_axes.plot([1, 2, 3], [3, 1, 4])
inset_axes.set_title('Gráfico da Inserção')

# Exibir os gráficos
plt.show()
````

Neste exemplo, criamos um gráfico principal usando `plt.plot()` e um gráfico de inserção separado usando `plt.axes()`. A função `plt.axes()` recebe quatro argumentos: a posição da inserção dentro do gráfico principal (esquerda, inferior, largura, altura). Ajustando esses valores, podemos posicionar a inserção no local desejado.

## 7. Plotagem 3D no Matplotlib


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

### Criando um Gráfico 3D Simples


````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gerar dados
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
print(Z)

# Criar o gráfico 3D
ax.plot_surface(X, Y, Z)

# Definir rótulos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar o gráfico
plt.show()
````

### Gráficos de Superfície
Gráficos de superfície são excelentes para visualizar funções de duas variáveis no espaço 3D. O Matplotlib fornece a função `plot_surface` para criar esses gráficos. Vamos explorar um gráfico de superfície mais complexo da função de Rosenbrock.


````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gerar dados
a = 1
b = 100
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = (a - X)**2 + b * (Y - X**2)**2

# Criar gráfico de superfície
ax.plot_surface(X, Y, Z, cmap='viridis')

# Definir rótulos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar o gráfico
plt.show()
````

### Gráficos de Dispersão (Scatter Plots)


````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gerar dados aleatórios
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Criar gráfico de dispersão
ax.scatter(x, y, z, c='r', marker='o')

# Definir rótulos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar o gráfico
plt.show()
````

### Gráficos em Wireframe

Gráficos em wireframe são úteis para visualizar superfícies em 3D. A função `plot_wireframe` do Matplotlib nos permite criar esses gráficos.


````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gerar dados
u = np.linspace(0, 2 * np.pi, 10)
v = np.linspace(0, np.pi, 10)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

# Criar gráfico em arame
ax.plot_wireframe(x, y, z)

# Definir rótulos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar o gráfico
plt.show()
````

### Gráficos de Contorno

Os gráficos de contorno são comumente usados para visualizar dados 3D em um plano 2D. As funções `contour` e `contourf` do Matplotlib nos permitem criar esses gráficos. Vamos plotar um mapa de contorno de uma função.


````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gerar dados
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Criar gráfico de contorno
ax.contourf(X, Y, Z, cmap='coolwarm')

# Definir rótulos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar o gráfico
plt.show()
````

### Personalizando Gráficos 3D

O Matplotlib oferece várias opções de personalização para aprimorar a representação visual de gráficos 3D. Algumas opções comuns incluem ajustar o ângulo de visualização, adicionar uma barra de cores, definir limites dos eixos e adicionar linhas de grade.


````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gerar dados
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Criar gráfico 3D
ax.plot_surface(X, Y, Z, cmap='viridis')

# Personalizar gráfico
ax.view_init(elev=30, azim=45)  # Definir o ângulo de visualização
fig.colorbar(ax.plot_surface(X, Y, Z, cmap='viridis'), ax=ax)  # Adicionar uma barra de cores
ax.set_xlim([-5, 5])  # Definir limites do eixo X
ax.set_ylim([-5, 5])  # Definir limites do eixo Y
ax.set_zlim([0, 50])  # Definir limites do eixo Z
ax.grid(True)  # Adicionar linhas de grade

# Definir rótulos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar o gráfico
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()
ax.set_xlim([0, 10])

scat = ax.scatter(1, 0)
x = np.linspace(0, 10)


def animate(i):
    scat.set_offsets((x[i], 0))
    return scat,

ani = animation.FuncAnimation(fig, animate, repeat=True,
                                    frames=len(x) - 1, interval=50)

# To save the animation using Pillow as a gif
writer = animation.PillowWriter(fps=15,
                                metadata=dict(artist='Me'),
                                bitrate=1800)
ani.save('scatter.gif', writer=writer)

plt.show()
````
