````python
import cv2
from matplotlib import pyplot as plt
import numpy as np

def imshow(image = None, size = 10):
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = h/w
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
````

### **Translações**

Transformação que desloca a posição de uma imagem.

Usamos o cv2.warpAffine para implementar essas transformações.

```cv2.warpAffine(imagem, T, (largura, altura))```

![](https://raw.githubusercontent.com/rajeevratan84/ModernComputerVision/main/warp.png)

````python
# Imagem de teste
image = cv2.imread('images/tobago.jpg')
imshow(image)

# Armazena a altura e a largura da imagem.
height, width = image.shape[:2]

# Vamos deslocá-la em um quarto da altura e da largura.
quarter_height, quarter_width = height/4, width/4

#       | 1 0 Tx |
#  T  = | 0 1 Ty |

T = np.float32([[1, 0, quarter_width], [0, 1,quarter_height]])

img_translation = cv2.warpAffine(image, T, (int(width), int(height)))

imshow(img_translation)
````

### **Rotação**

```cv2.getRotationMatrix2D(rotation_center_x, rotation_center_y, angle of rotation, scale)```

![](https://github.com/rajeevratan84/ModernComputerVision/raw/main/rotation.png)


````python
# Imagem de teste
image = cv2.imread('images/tobago.jpg')
height, width = image.shape[:2]

# Divida por dois para rotacionar a imagem em torno do seu centro.
rotation_matrix = cv2.getRotationMatrix2D((width/3, height/2), 45, 0.5)
print(rotation_matrix)

# Insira nossa imagem, a matriz de rotação e a largura e altura final desejadas.
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

imshow(rotated_image)
````

````python
# Flip.
flipped = cv2.flip(image, 1)
imshow(flipped)
````
